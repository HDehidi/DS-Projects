```python
def larger_diagonals(A,B):
    aDiag = np.diag(A)
    bDiag = np.diag(B)
    
    aAvg = np.mean(aDiag)
    bAvg = np.mean(bDiag)
    
    if aAvg > bAvg:
        return A
    else:
        return B
```

```python
A = np.zeros((4,3)) + 3
B = np.eye(5)
larger_diagonals(A,B)
```

```python
def add_columns(A):
    return A[:,::2] + A[:,1::2]
```

```python
A = np.zeros((3,4))
A[:,1::2] = np.ones((3,2))*-1
A[:,::2] = np.ones((3,2))*2
print(A)
```

```python
add_columns(A)
```

```python
import math
def create_checkerboard(n):
    checkerboard = np.zeros((n,n))
    checkerboard[::2,1::2] = np.ones((math.ceil(n/2),n//2))
    checkerboard[1::2,::2] = np.ones((n//2,math.ceil(n/2)))
    return checkerboard
```

```python
create_checkerboard(6)
```

```python
create_checkerboard(7)
```

```python
def create_checkerboardNoCeil(n):
    checkerboard = np.zeros((n,n))
    if n%2 == 0:
        checkerboard[::2,1::2] = np.ones((n//2,n//2))
        checkerboard[1::2,::2] = np.ones((n//2,n//2))
    else:
        checkerboard[::2,1::2] = np.ones((n//2+1,n//2))
        checkerboard[1::2,::2] = np.ones((n//2,n//2+1))
    return checkerboard
```
